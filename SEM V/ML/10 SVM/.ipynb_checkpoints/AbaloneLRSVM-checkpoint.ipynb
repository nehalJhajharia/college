```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('abalone.csv')
df.head(10)
```

## EDA

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
# Check for null
df.isna().any()
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr())
```

```python
X = df.drop('Sex', axis=1)
X
```

```python
y = df['Sex']
y
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaledX = scaler.fit_transform(X)
scaledX
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
sns.heatmap(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

## SVM

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(X_train, y_train)
pred2 = svc.predict(X_test)
```

```python
accuracy_score(y_test, pred2)
```

```python
cm = confusion_matrix(y_test, pred2)
cm
```

```python
sns.heatmap(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred2))
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
pca.singular_values_
```

```python
X_new = pca.transform(X)
X_new
```

```python
X_PCA = pd.DataFrame(X_new)
X_PCA
```

```python
#sns.heatmap(X_PCA.corr()) 
```

Logistic after PCA

```python
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(X_PCA, y, test_size=0.2)
```

```python
clf_PCA = LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
clf_PCA.fit(X_train_PCA, y_train_PCA)
pred_PCA = clf_PCA.predict(X_test_PCA)
```

```python
accuracy_score(y_test_PCA, pred_PCA)
```

```python
cm_PCA = confusion_matrix(y_test_PCA, pred_PCA)
cm_PCA
```

```python
sns.heatmap(cm_PCA)
```

```python
print(classification_report(y_test_PCA, pred_PCA))
```

SVM after PCA

```python
svc_PCA = SVC()
```

```python
svc_PCA.fit(X_train_PCA, y_train_PCA)
pred2_PCA = svc_PCA.predict(X_test_PCA)
```

```python
accuracy_score(y_test_PCA, pred2_PCA)
```

```python
cm_PCA = confusion_matrix(y_test_PCA, pred2_PCA)
cm_PCA
```

```python
sns.heatmap(cm_PCA)
```

```python
print(classification_report(y_test_PCA, pred2_PCA))
```
