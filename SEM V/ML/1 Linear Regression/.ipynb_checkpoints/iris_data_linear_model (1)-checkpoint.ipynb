```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns', None);
```

```python
df = pd.read_csv('iris.csv')
print(df)
```

```python
# datasets into features and targets
y = pd.DataFrame()
y['Sepal.Length'] = df['Sepal.Length']
x = df.drop('Sepal.Length', axis = 1)
print(x.head())
print(y.head())
```

```python
x.corr()
```

```python
sns.pairplot(x)
```

```python
plt.xlabel("Petal.Length")
plt.ylabel("Petal.Width")
plt.title("Petal.Length VS Petal.Width")
plt.scatter(x["Petal.Length"], x["Petal.Width"])
plt.show()
```

```python
# Correlation between dependent and each independent variable
for feature in x:
    print("Correlation ", feature, "with sepal length : \n", np.corrcoef(y['Sepal.Length'], x[feature]))
```

```python
for feature in x.columns:
    if(x[feature].dtype == 'O'):
        continue
    data_copy = x.copy()
    plt.title(feature)
    x.boxplot(column = feature)
    plt.show()
```

```python
differ_species = [labels for labels in df['Species']]
differ_species_set = set(differ_species)
differ_species_set
```

```python
# Converting Categorical featues
dummies = pd.get_dummies(x.Species)
print(dummies)
x = x.drop('Species', axis = 1)
print(x.head())
```

```python
# concatenating 
x = pd.concat([x, dummies], axis = 'columns')
x.head()
```

```python
# Checking MultiCollinearity for independent variables or features with VIF
# from statsmodels.stats.outliers_influence import variance_inflation_factor
# vif_data = pd.DataFrame()
# vif_data["features"] = x.columns
# vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
# print(vif_data)
```

```python
# Dividing into training and test datasets
x_train = x[:-30]
y_train = y[:-30]
x_test = x[-30:]
y_test = y[-30:]
```

```python
print(x_train.head())
print(y_train.head())
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(x_train, y_train)
```

```python
# Making Predictions
predictions = regr.predict(x_test)
print(predictions)
print(y_test)
a = pd.DataFrame()
a['y'] = y_test
a['predi'] = predictions
print(a)
```

```python
# Calculating Mean Squared Error
print("Mean Squared Error", mean_squared_error(y_test, predictions))
```

```python
# Coefficient of determination if 1 then perfect model
print("Coefficient of Determination", r2_score(y_test, predictions))
```
